Carga el tiempo meteorológico de Praga en un data.frame. Fichero: *EZE00100082*

Mostrar los 10 años que más precipitiaciones han tenido junto con su valor correspondiente (columna PRCP):

* Muestra en un diagrama de dispersión la temperatura mínima registrada en un año (columna EMNT) en función del año registrado (columna DATE).
* Crear una regresión lineal (polinomio orden 1) que muestre la tendencia. Pintarla sobre el diagrama de dispersión.
* En media, ¿Cuantos grados ha aumentado la temperatura mínima cada año? (pendiente de la  regresión lineal)

```r
library("ggplot2")

model<-lm(data=weatherPraga, formula = EMNT ~ DATE)
#summary(model)
paste("La pendiente de la regresion lineal es",model$coefficients[2])

weatherPraga$predicted<-predict(model,weatherPraga)

ggplot(data=weatherPraga,aes(x=DATE,y=EMNT))+geom_point(size=1)+geom_line(aes(y=predicted),color="blue")+
                theme_bw()+xlab("Anyo registrado")+ylab("Temperatura minima")
```

¿Ha subido realmente la temperatura mínima?
* Calcula la media de la temperatura mínima (columna EMNT) del intervalo de años [1775,1794]
* Calcula la media de la temperatura mínima (columna EMNT) del intervalo de años [1985,2004]
* ¿Cual es la diferencia entre ambos? ¿Es estadísticamente significativa?
  * Nota: antes de aplicar un test estadístico hay que ver si cumple con los requisitos de ese test.

```r
intervalo<-data.frame(weatherPraga$EMNT,weatherPraga$DATE)

intervalo1<-intervalo[1:20,]

paste("La media del primer intervalo es",mean(intervalo1$weatherPraga.EMNT))

intervalo2<-intervalo[(nrow(intervalo)-19):nrow(intervalo),]

paste("La media del segundo intervalo es",mean(intervalo2$weatherPraga.EMNT))
```

```r
#Vamos a comprobar si la diferencia de las medias de ambos intervalos ha cambiado con un test t-student emparejado.
#Para ello debemos comprobar en primer lugar si estas distribuciones son gaussianas con el test de Shapiro.

shapiro.test(intervalo1$weatherPraga.EMNT)
shapiro.test(intervalo2$weatherPraga.EMNT)
```

```r
#En el primer caso, la distribución es asimilable a una normal ya que el p-valor es razonablemente alto.
#En el segundo caso, estamos en el límite. Vamos a pintar la distribución y compararla con una gaussiana.

library(ggplot2)
ggplot(data=intervalo2,aes(x=weatherPraga.EMNT))+
    stat_function(fun=dnorm,args = list(mean = mean(intervalo2$weatherPraga.EMNT), sd=sd(intervalo2$weatherPraga.EMNT)), geom="area",color="green",fill="green",alpha=0.1)+
    geom_density(color="red")+
    theme_linedraw()
```

Podemos comprobar como la distribución tiene una forma similar a la gaussiana salvo por dos aspectos:
* Por un lado, está desplazada a la derecha. Parece que las temperaturas minimas más altas han presentado un número de ocurrencias mayor que las más bajas en los últimos 20 años del registro.
* Por otro, el extremo inferior tiene una probabilidad de ocurrencia muy alto.

Al tratarse de una serie corta, asumimos que estos aspectos se deben a la alatoriedad de las temperaturas y que en una serie mayor efectivamente se tendería a una distibución gaussiana (como de hecho tiende la serie del otro intervalo).
De cara al ejercicio supondremos que la distribución se asemeja auna gaussiana (p-valor>0,05) y aplicamos el test t-student emparejado:

```r
t.test(intervalo1$weatherPraga.EMNT,intervalo2$weatherPraga.EMNT,paired=TRUE)
```

Al obtener un p-valor muy bajo (<0,05) podemos descartar la hipótesis 0: Que las medias son iguales. 
Por lo tanto, la media de temperaturas mínimas ha cambiado, siendo la media de las diferencias de casi 4 grados.

```r
regions<-c(which(regionData$X...Country.Code=="WLD"),
    which(regionData$X...Country.Code=="EUU"),
    which(regionData$X...Country.Code=="SSA"),
    which(regionData$X...Country.Code=="NAC"),
    which(regionData$X...Country.Code=="TEA"))

newRegions<-data.frame(Country.Code=regionData$X...Country.Code[regions])
newRegionData<-merge(newRegions,databank,by="Country.Code", all.x = TRUE)

CO2Totals<-subset(newRegionData,Indicator.Code=="EN.ATM.GHGT.KT.CE")

require(reshape2)

CO2Melt<-melt(CO2Totals,na.rm=TRUE)
```

Suponiendo que la tendencia se mantiene, es decir, que no se aplica ninguna política a nivel global de reducción de emisiones de CO2, ¿cual sería el total de emisiones de CO2 en todo el mundo en el año 2020?

Para calcularlo utiliza tres modelos lineales:
* Regresión lineal con polinomio grado 1
* Regresión lineal polinomio grado 3
* Regresión lineal polinomio grado 10

Represénta en una gráfica el valor real y el calculado por cada uno de los 3 modelos.

```r
CO2Melt$variable <- as.numeric(CO2Melt$variable)
CO2Model<-subset(CO2Melt,Country.Code=="WLD")

model2<-lm(data=CO2Model, formula = value/1000 ~ variable)
#summary(model2)
CO2Model$predicted2<-predict(model2,CO2Model)

model3<-lm(data=CO2Model, formula = value/1000~poly(variable,3))
#summary(model3)
CO2Model$predicted3<-predict(model3,CO2Model)

model10<-lm(data=CO2Model, formula = value/1000~poly(variable,10))
#summary(model10)
CO2Model$predicted10<-predict(model10,CO2Model)

ggplot(data=CO2Model,aes(x=variable,y=value/1000))+geom_point(size=1)+geom_line(aes(y=predicted2),color="blue")+
        geom_line(aes(y=predicted3),color="green")+
        geom_line(aes(y=predicted10),color="red")+
                xlab("Anyos (1970-2012)")+ylab("Emisiones CO2(kt) mundiales")+ggtitle("Regresion lineal grados 1, 3 y 10")+
                  theme(panel.background = element_blank(),
                    axis.text.x=element_blank(),
                    axis.ticks.x=element_blank())

```

```r
m<-round(model2$coefficients[2],2)
b<-round(model2$coefficients[1],2)
a<-round(predict(model2,CO2Model)[nrow(CO2Model)],2)
a2012<-round((a-b)/m,0)
a2020<-a2012+8
paste("Sabemos que y=",a,"en el anyo 2012. Por lo tanto en el 2012 x=",
     a2012)
paste("En el anyo 2020 x=",a2020)
paste("El total de emisiones de CO2 en 2020 seria",
      m,"*",a2020,"+",b, "kt:")
print(round(m*a2020+b,2))
```

## Nivel de CO2

En esta parte vamos a trabajar con el nivel medio de CO2 medido por el NOAA en Manuna Loa (Hawai) desde 1968.

```r
co2ppm<-read.csv("./datos/co2_mm_mlo.txt",skip=73,sep="",header = F, 
                 col.names= c("year","month","date","average","interpolated","trend","days"))
```

Representa el diagrama de dispersión entre:
* CO2 medido en la atmósfera (co2ppm\$interpolated)
* CO2 emitido por todos los países (databank con Indicator.Code=="EN.ATM.CO2E.KT" y Country.Code=="WLD"?

¿Cual es su correlación?


Nota: 

* Los valores en co2ppm vienen en meses, habrá que agregarlos, calcular la media anual a partir de los meses para así poder hacer la correlación con databank.
* Los valores en databank vienen en columnas, habrá que pasarlos a filas utilizando la función *melt*.

```r
co2ppmAnual<-c()
for (i in 1958:2014){

    iter<-subset(co2ppm,year==toString(i))
    co2ppmAnual <- c(co2ppmAnual, mean(iter$interpolated))
    }

co2ppmAnual<-data.frame("years"=1958:2014, "value"=co2ppmAnual)

ggplot(data=co2ppmAnual,aes(x=years,y=value))+geom_point(color="red",size=1)+
                xlab("Anyos (1958-2014)")+ylab("Media anual CO2")+
                ggtitle("CO2 medido en atmosfera")+
                  theme_bw()
```

```r
CO2Totals2<-subset(newRegionData,Indicator.Code=="EN.ATM.CO2E.KT")

require(reshape2)

CO2Melt2<-melt(CO2Totals2,na.rm=TRUE)
CO2Corr<-subset(CO2Melt2,Country.Code=="WLD")

CO2Corr<-data.frame("years"=1960:2014, "value"=CO2Corr$value)

ggplot(data=CO2Corr,aes(x=years,y=value))+geom_point(color="blue",size=1)+
                xlab("Anyos (1960-2014)")+ylab("Emisiones CO2(kt)")+
                ggtitle("Emisiones CO2(kt) mundiales")+
                  theme_bw()
```

```r
paste("La correlacion entre ambas variables es:",cor(co2ppmAnual$value[3:nrow(co2ppmAnual)],CO2Corr$value))
```
